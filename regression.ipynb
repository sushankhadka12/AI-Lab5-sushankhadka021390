```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = [
    ["Aayush", 23, 32000],
    ["Bikash", 25, 40000],
    ["Sita", 28, 47000],
    ["Ram", 30, 52000],
    ["Hari", 35, 61000],
    ["Gita", 40, 75000],
    ["Nabin", 38, 71000],
    ["Sabin", 34, 68000],
    ["Sunita", 36, 69000],
    ["Ramesh", 45, 85000],
    ["Rekha", 50, 92000],
    ["Suman", 48, 89000],
    ["Manish", 42, 77000],
    ["Anjali", 33, 65000],
    ["Kabita", 27, 45000],
]

df = pd.DataFrame(data, columns=["Name", "Age", "Income($)"])

X = df["Age"].values
y = df["Income($)"].values

x_mean = np.mean(X)
y_mean = np.mean(y)

numerator = np.sum((X - x_mean) * (y - y_mean))
denominator = np.sum((X - x_mean) ** 2)
w = numerator / denominator
b = y_mean - w * x_mean

y_pred = w * X + b

plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='blue', label='Actual Income')
plt.plot(X[np.argsort(X)], y_pred[np.argsort(X)], color='red', label='Predicted Line')
plt.xlabel("Age")
plt.ylabel("Income ($)")
plt.title("Linear Regression: Age vs Income")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print(f"Linear Regression Equation: Income = {w:.2f} * Age + {b:.2f}")
```
